# [Annexe] Modélisation du géopotentiel

Lors de la modélisation de l'influence du potentiel gravitationnel de la Terre, l'approche la plus simple est de considérer la planète comme une sphère parfaite de masse $M$, ou comme un particule de même masse qui serait située au centre de cette sphère, ce qui est équivalent. Toutefois, cette hypothèse est très simplificatrice : la Terre n'est pas une sphère parfaite (elle s'approche d'une ellipsoïde oblate), présente des variations locales de masse, est impactée par sa propre rotation, et cette liste est loin d'être exaustive.

En conséquence, le champ de gravitation terrestre n'est pas uniforme. Nous introduisons donc la notion de **géopotentiel** qui désigne le potentiel gravitationnel en un point autour de la Terre et à une altitude donnée.

## Formule générale

Pour prendre en compte que la masse de la Terre n'est pas uniformément répartie dans une sphère parfaite, nous intégrons le potentiel $U$ de la gravité terrestre pour modéliser sa force d'attraction sur un satellite de son voisinage. Nous avons :

$$\mathbf{\ddot r} = \nabla U$$
avec $U = GM\dfrac1r$, où $G$ est la constante universelle de gravitation et $M$ la masse de la Terre.

Nous généralisons l'expression à une distribution de masse arbitraire $\rho(s)\,d^3\mathbf s$ (où ${\rho(s)}$ est la densité à un point $s$ au sein de la Terre) :

$$U = \int \frac{\rho(s)\,d^3\mathbf s}{|\mathbf r - \mathbf s|}$$
avec $|\mathbf r - \mathbf s|$ la distance entre le point $s$ et le satellite.


## Polynômes de Legendre

Pour calculer l'intégrale, le dénominateur de l'équation précénte, se développe en une série de polynômes de Legendre (étant admis $r>s$) :

$$
\frac{1}{|\mathbf r - \mathbf s|} = \frac{1}{r} \sum_{n=0}^\infty \left(\frac{s}{r}\right)^n - P_n\left(\cos\gamma\right)
$$
avec $\gamma$ l'angle formé par $\mathbf r$ et $\mathbf s$, impliquant $\cos\gamma = \dfrac{\mathbf r \cdot \mathbf s}{r \ s}$, et $P_n(u)$ est le polynôme de Legendre $P_n$ de degré $n$, tel que :

$$P_n(u)=\frac1{2^nn!}\frac{\mathrm d^n}{\mathrm du^n}\left(u^2-1\right)^n$$

En introduisant $\lambda, \phi$ les longitude et latitude de $r$, et $\lambda',\phi'$ celles de $s$, le théorème des polynome de Legendre s'écrit :

$$
P_n\left(\cos\gamma\right) = \sum_{m=0}^{n} \left(2 - \delta_{0m}\right) \frac{(n-m)!}{(n+m)!} P_{nm}\left(\sin\phi\right) P_{nm}\left(\sin\phi'\right)\cos\left(m(\lambda - \lambda')\right)
$$

avec  et $\delta_{0m}$ le symbole de Kronecker appliqué à $0,m$, valant donc :

$$
\delta_{0m} = \begin{cases}1&\text{si }m=0\\0&\text{si }m\ne0\end{cases}
$$

Avec $P_{nm}$ le polynôme de Legendre associé de degré $n$ et d'ordre $m$, définit comme :

$$
P_{nm}(u) = \left(1 - u^2\right)^{\frac{m}2}\frac{\mathrm d^m}{\mathrm du^m}P_n(u)
$$

Les expressions des polynomes de Legendre des premiers degrés et ordres sont fournies par *Satellite Orbits* de Montenbruck  et Gill [[1](#ref-1)] (p.57) :

| $n$ | $m$ | $P_{nm}(u)$                  | $P_{nm}(\sin\phi)$                  |
|:---:|:---:|:----------------------------:|:-----------------------------------:|
| $0$ | $0$ | $1$                          | $1$                                 |
| $1$ | $0$ | $u$                          | $\sin\phi$                          |
| $1$ | $1$ | $\sqrt{1-u^2}$               | $\cos\phi$                          |
| $2$ | $0$ | $\frac12\left(3u^2-1\right)$ | $\frac12\left(3\sin^2\phi-1\right)$ |
| $2$ | $1$ | $3u\sqrt{1-u^2}$             | $3\cos\phi\sin\phi$                 |
| $2$ | $2$ | $3\left(1-u^2\right)$        | $3\cos^2\phi$                       |




## Application au géopotentiel

Le développement de la formule du géopotentiel est donc :

$$
U = \frac{GM}r \sum_{n=0}^\infty \sum_{m=0}^n \frac{R^n}{r^n} P_{nm}\left(\sin\phi\right)\big(C_{nm} \cos(m\lambda) + S_{nm} \sin(m\lambda)\big)
$$

avec $R$ le rayon moyen volumétrique terrestre, et où les coefficients $C_{nm}$ et $S_{nm}$ décrivent la dépendance à la distribution de masse interne de la Terre, tels que :

$$
\begin{align*}
C_{nm} = \frac{2-\delta_{0m}}M \frac{(n-m)!}{(n+m)!} \int\frac{s^n}{R^n} P_{nm}\!\left(\sin\phi'\right)\cos\left(m\lambda'\right)\rho(s)d^3s\\
S_{nm} = \frac{2-\delta_{0m}}M \frac{(n-m)!}{(n+m)!} \int\frac{s^n}{R^n} P_{nm}\!\left(\sin\phi'\right)\sin\left(m\lambda'\right)\rho(s)d^3s
\end{align*}
$$

Ces coefficients sont normalisés tels que :

$$
\begin{align*}
\bar{P}_{nm} &= \sqrt \frac{(2-\delta_{0m})(2n+1)(n-m)!}{(n+m)!} P_{nm} \\
\begin{Bmatrix}\bar C_{nm}\\\bar S_{nm}\end{Bmatrix} &=
\sqrt\frac{\left(n+m\right)!}{\left(2-\delta_{0m}\right)\left(2n+1\right)\left(n-m\right)!}
\begin{Bmatrix}C_{nm}\\S_{nm}\end{Bmatrix}
\end{align*}
$$

Nous en déduisons l'accélération subit par le satellite du fait du géopotentiel :

$$
\boxed{\mathbf{\ddot r} = \nabla\frac{GM}{\mathbf r} \sum_{n=0}^\infty \sum_{m=0}^n \frac{R^n}{r^n} \bar P_{nm}\!\left(\sin\phi\right)\left(\bar C_{nm} \cos(m\lambda) + \bar S_{nm} \sin(m\lambda)\right)}
$$

où chaque combinaison de $n,m$ est représenté sous la forme $J_{n,m}$ :

$$\mathbf{\ddot r} = \nabla\frac{GM}{\mathbf r} \sum_{n=0}^\infty \sum_{m=0}^n J_{n,m}$$

## Coefficients $J_{n,m}$ du géopotentiel

### Coefficients zonaux

Avec $m=0$, les coefficients obtenus sont les coefficients zonaux. Ils décrivent la partie du potentiel qui ne dépend pas de la longitude. Les termes en $S_{n0}$ s'annulent alors.

#### Usage des coefficients
* $J_0=1$ et décrit la contribution de la sphère au géopotentiel,
* $J_1$ est nul, conséquence de la position du centre de masse de la Terre au centre du référentiel.
* $J_2$ décrit la contribution du caractère oblate (légèrement aplati) de la Terre.
* Les coefficients suivants sont plus de trois odres de grandeurs inférieurs à $J_2$, et ne sont pour cette raison pas inclus dans notre modélisation.

### Autres coefficients

Les coefficients pour lesquels $m>0$ dépendent de la longitude. Ils sont appelés tesséraux pour $m<n$ et sectoriaux pour $m=n$.

## Calcul des cofficients $J_{n,m}$

De la définition en série de l'accélération précédement introduite, et de la limite de précision que vous avons définie, nous devons déterminer les cofficients $J_{n,m}$ suivants :
$$
\mathbf{\ddot r} = \nabla\frac{GM}{\mathbf r} \sum_{n=0}^2 \sum_{m=0}^n J_{n,m}
 = \nabla\frac{GM}{\mathbf r} \left(J_{0,0} + J_{1,0} + J_{1,1} + J_{2,0} + J_{2,1} + J_{2,2}\right)
$$

Nous nous référons aux valeurs fournies par *Satellite Orbits* [[1](#ref-1)] (p.60) pour les valeurs de $C_{nm}$ et $S_{nm}$ :

| $1$       | $\phantom{-}0.00$                       | $\phantom{-}0.00$   |                     |                     |

Des valeurs nulles de la table ci-dessus, nous déduisons immédiatement $J_{1,0} = J_{1,1} = J_{2,1} = 0$.

Ayant précédement introduit $J_{0,0} = 1$, nous avons donc :

$$\mathbf{\ddot r} = \nabla\frac{GM}{\mathbf r} \left(1 + J_{2,0} + J_{2,2}\right)$$

### Calcul de $J_{2,0}$
De la définition du coefficient :
$$J_2 = \frac{R^2}{r^2} \bar P_{2,0}\!\left(\sin\phi\right) \bar C_{2,0} $$

de la formule des coefficient normalisés :
$$\bar{P}_{2,0} = \sqrt \frac{5(2-\delta_{00})\times2!}{2!} P_{2,0} = \sqrt{5} P_{2,0}$$

et de la valeur $P_{2,0}\!\left(\sin\phi\right) = \frac12\left(3\sin^2\phi - 1\right)$ issue de la Table 1, on trouve :

$$\boxed{J_2 = \frac{R^2}{r^2} \sqrt{5}  \frac12\left(3\sin^2\phi - 1\right) \bar C_{2,0}}$$

### Calcul de $J_{2,2}$
De la définition du coefficient :
$$J_{2,2}=\frac{R^2}{r^2}\bar P_{2,2}\!\left(\sin\phi\right)\left(\bar C_{2,2}\cos(m\lambda)+\bar S_{2,2}\sin(m\lambda)\right)$$
de la formule des coefficient normalisés :
$$\bar P_{2,2} = \sqrt\frac{5(2-\delta_{02})}{4!} P_{2,2} = \sqrt\frac{10}{24} P_{2,2} = \frac{\sqrt{15}}6 P_{2,2}$$

et de la valeur $P_{2,2}\!\left(\sin\phi\right) =3\cos^2\phi$ issue de la Table 1, on trouve :
$$\boxed{J_{2,2}=\frac{R^2}{r^2}\frac{\sqrt{15}}2\cos^2\phi\left(\bar C_{2,2}\cos(2\lambda)+\bar S_{2,2}\sin(2\lambda)\right)}$$

## Gradient de la fonction scalaire associée $\nabla\frac{GM}{\mathbf r}$

Le gradient d'une fonction scalaire $f(x,y,z)$ en coordonnées cartésiennes est donné par :

$$\nabla f = \frac{\partial f}{\partial x} \mathbf i + \frac{\partial f}{\partial y} \mathbf j + \frac{\partial f}{\partial z} \mathbf k$$

où $\mathbf{i}$, $\mathbf{j}$ et $\mathbf{k}$ sont les vecteurs unitaires dans les directions $x$, $y$ et $z$ respectivement.

Pour la position d'un satellite dans le repère cartésien géocentrique, on a alors :

$$
\begin{align*}
\nabla\frac{GM}{\mathbf r}
&= GM\left[
       \frac{\mathrm d}{\mathrm dx}\left(\frac1{\sqrt{x^2+y^2+z^2}}\right)
       +\frac{\mathrm d}{\mathrm dy}\left(\frac1{\sqrt{x^2+y^2+z^2}}\right)
       +\frac{\mathrm d}{\mathrm dz}\left(\frac1{\sqrt{x^2+y^2+z^2}}\right)
     \right]\\
&= GM\left[
       -\frac{\mathbf x}{\left(x^2+y^2+z^2\right)^{\frac32}}
       -\frac{\mathbf y}{\left(x^2+y^2+z^2\right)^{\frac32}}
       -\frac{\mathbf z}{\left(x^2+y^2+z^2\right)^{\frac32}}
     \right]\\
&= -GM\left(
       \frac{\mathbf x+\mathbf y+\mathbf z}{r^3}
     \right)\\
&= \boxed{- \frac{GM}{r^3}\mathbf r}
\end{align*}
$$

## Formule de l'accélération géopotentielle

Nous obtenons donc, pour un calcul du géopotentiel jusqu'à $n=2$ en un point $(r,\lambda,\phi)$ dans le système de coordonnées sphériques en convension rayon-longitude-latitude :

$$
\begin{align*}
\mathbf{\ddot r} &= \nabla\frac{GM}{\mathbf r} \left(J_0 + J_{2,0} + J_{2,2}\right) \\
&= - \frac{GM}{r^3} \left(1
  + \frac{R^2}{r^2} \sqrt{5}  \frac12\left(3\sin^2\phi - 1\right) \bar C_{2,0}
  + \frac{R^2}{r^2}\frac{\sqrt{15}}2\cos^2\phi\left(\bar C_{2,2}\cos(2\lambda)+\bar S_{2,2}\sin(2\lambda)\right)
\right)\mathbf r\\
&=\boxed{- \frac{GM}{r^3} \left[1
  + \frac{R^2}{r^2}\frac{\sqrt{5}}2\Bigg(
    \left(3\sin^2\phi - 1\right) \bar C_{2,0}
    +\sqrt{3}\cos^2\phi\left(\bar C_{2,2}\cos(2\lambda)+\bar S_{2,2}\sin(2\lambda)\right)
\Bigg)\right]\mathbf r}
\end{align*}
$$

### Application numériques

À partir des valeurs des coefficients $C_{2,0}$, $C_{2,2}$ et $S_{2,2}$ issues de la Table 2, nous calculons les valeurs des coefficients normalisés à utiliser en application numérique :

$$
\begin{align*}
\bar C_{2,0} &= \sqrt\frac15 C_{2,0} = \frac{\sqrt5}5 \times -1.08\cdot10^{-3} \approx -4.83\cdot10^{-4}\\
\bar C_{2,2} &= \sqrt\frac{24}{10} C_{2,2} = \frac{2\sqrt{15}}5 \times 1.57\cdot10^{-6} \approx 2.43\cdot10^{-6}\\
\bar S_{2,2} &= \sqrt\frac{24}{10} S_{2,2} = \frac{2\sqrt{15}}5 \times -9.03\cdot10^{-7} \approx -1.40\cdot10^{-6}
\end{align*}
$$
